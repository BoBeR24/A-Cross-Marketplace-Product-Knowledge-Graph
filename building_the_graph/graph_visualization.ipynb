```python
from rdflib import Graph, Literal, URIRef
from rdflib.extras.external_graph_libs import rdflib_to_networkx_multidigraph
from pyvis.network import Network
import random
```

```python
g = Graph().parse("../data/ontology_graph.ttl")
```

```python
# 2. Convert to a NetworkX MultiDiGraph
# This preserves the direction and multiple edges between nodes
G = rdflib_to_networkx_multidigraph(g)


# 3. Create a Pyvis Network object
net = Network(height="1080px", width="100%", bgcolor="#222222", font_color="white", directed=True)

# 4. Load the NetworkX graph into Pyvis
net.from_nx(G)

# Optional: Add physics buttons so you can tweak the layout live in the browser


# 5. Save and open in your browser
# net.show("rdf_graph.html", notebook=False)

html = net.generate_html()

output_html = "graph_pyvis.html"

with open(output_html, "w") as f:
    f.write(html)
```
