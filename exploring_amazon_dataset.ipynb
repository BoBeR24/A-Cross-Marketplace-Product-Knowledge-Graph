```python
import re
```

```python
def clean_text(text):

    # 1. This regex keeps: a-z, A-Z, 0-9, and standard punctuation
    # It removes everything else (including emojis and CJK brackets)
    cleaned = re.sub(r'[^ -~]', '', text)

    # 2. Optional: Strip extra whitespace left behind
    return cleaned.strip()
```
